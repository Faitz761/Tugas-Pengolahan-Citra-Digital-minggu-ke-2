# **NOMOR 2**



```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/content/kucing.jpg', cv2.IMREAD_GRAYSCALE)

sizes = [512, 256, 128, 64]

for s in sizes:
    resized = cv2.resize(img, (s, s), interpolation=cv2.INTER_AREA)
    plt.imshow(resized, cmap='gray')
    plt.title(f'Resize {s}x{s}')
    plt.axis('off')
    plt.show()
```

# **NOMOR 3**



```python
img = cv2.imread('/content/kucing.jpg', cv2.IMREAD_GRAYSCALE)

img_2bit = (img // 64) * 64
img_4bit = (img // 16) * 16
img_8bit = img

titles = ['2 Bit (4 Level)', '4 Bit (16 Level)', '8 Bit (256 Level)']
images = [img_2bit, img_4bit, img_8bit]

for i in range(3):
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
    plt.show()
```
