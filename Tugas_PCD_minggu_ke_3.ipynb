##**JAWABAN 1**

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/content/kucing.jpg', cv2.IMREAD_GRAYSCALE)

negative = 255 - img

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Citra Asli')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(negative, cmap='gray')
plt.title('Negative Image')
plt.axis('off')

plt.show()
```

##**JAWABAN 2**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/kucing.jpg', cv2.IMREAD_GRAYSCALE)

# ubah ke float
img_float = img.astype(np.float32)

# log transform
log_transform = np.log(1 + img_float)

# normalisasi
log_transform = log_transform / np.max(log_transform) * 255

# kembali ke uint8
log_transform = log_transform.astype(np.uint8)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Citra Asli')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(log_transform, cmap='gray')
plt.title('Log Transformation')
plt.axis('off')

plt.show()
```

##**JAWABAN 3**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/kucing.jpg', cv2.IMREAD_GRAYSCALE)

gamma = 0.5

normalized = img / 255.0
power_law = np.power(normalized, gamma)

power_law = np.uint8(power_law * 255)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Citra Asli')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(power_law, cmap='gray')
plt.title('Power Law (Gamma Correction)')
plt.axis('off')

plt.show()
```
